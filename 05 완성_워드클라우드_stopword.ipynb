## 불용어 처리 (1)

```python
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
# 한글 폰트 설정
plt.rcParams['font.family'] = 'Malgun Gothic'
# 유니코드에서  음수 부호설정
plt.rc('axes', unicode_minus=False)
```

```python
from konlpy.tag import Okt
import pandas as pd
#워드 클라우드
from wordcloud import WordCloud

okt = Okt()

df=pd.read_csv('./최종_전세사기_all_.csv')
text=' '.join(df['title']) # 제목만 추출
nouns=okt.nouns(text) # 제목에서 명사 추출
re_nouns=' '.join(nouns)
```

```python
re_nouns
```

```python
# 제목에서 명사추출한 리스트 pickle로 저장하기(re1)
import pickle
with open('title_nouns.pickle', 'wb') as f:
    pickle.dump(re_nouns, f, pickle.HIGHEST_PROTOCOL)

# load
# with open('nouns.pickle', 'rb') as f:
#     data = pickle.load(f)
```

```python
# stopwords 처리 전
wc = WordCloud("C:/windows/fonts/malgun.ttf",
               background_color = "white",
               max_words = 2000)
               #stopwords = stopwords)
wc = wc.generate(re_nouns) # str자료만 가능, list자료는 안됨
plt.figure(figsize = (10, 10))
plt.imshow(wc) #interpolation = "bilinear")
plt.axis("off")
plt.savefig('./제목만 워드클라우드.png')
plt.show()
```

### - 불용어 처리

```python
#stopwords 처리 후
stopwords=['전세','사기', '관련','사항']

wc = WordCloud("C:/windows/fonts/malgun.ttf",
               background_color = "white",
               max_words = 2000,
               stopwords = stopwords)
wc = wc.generate(re_nouns) # str자료만 가능, list자료는 안됨
plt.figure(figsize = (12, 12))
plt.imshow(wc) #interpolation = "bilinear")
plt.axis("off")
plt.savefig('./제목워드클라우드_stopword적용.jpg')
plt.show()
```

## 불용어 처리 (2)

```python
from collections import Counter
from wordcloud import WordCloud
import matplotlib.pyplot as plt
okt = Okt()

df=pd.read_csv('./최종_전세사기_all_.csv')
text=' '.join(df['title']) # 제목만 추출
nouns=okt.nouns(text) # 제목에서 명사 추출
nouns
```

```python
#단어 빈도수 세기
counter = Counter(nouns)
counter
```

```python
wc = WordCloud("C:/windows/fonts/malgun.ttf",
               background_color = "white",
               max_words = 2000)
wc = wc.generate_from_frequencies(counter) # str자료만 가능, list자료는 안됨
plt.figure(figsize = (12, 12))
plt.imshow(wc) #interpolation = "bilinear")
plt.axis("off")
plt.savefig('./제목워드클라우드_stopword적용.jpg')
plt.show()
```

```python
#stopword
stopword=['관련','문','질문','문의','시','사항','주','일','것','수','거','전세','전세계약']
nouns=[i for i in nouns if i not in stopword]
counter = Counter(nouns)
wc = WordCloud("C:/windows/fonts/malgun.ttf",
               background_color = "white",
               max_words = 2000)
wc = wc.generate_from_frequencies(counter) # str자료만 가능, list자료는 안됨
plt.figure(figsize = (12, 12))
plt.imshow(wc) #interpolation = "bilinear")
plt.axis("off")
plt.savefig('./제목워드클라우드_stopword적용.jpg')
plt.show()
```

```python
from collections import Counter
top=50
top_=counter.most_common(top)
top_
```

```python
df_top=pd.DataFrame(top_, columns=['word','freq']).set_index('word')
df_top.head()
```

```python
df_top.plot(kind='bar',figsize=(15,8))
plt.savefig('./제목freq50_bar_chart_전세사기단어_포함.png')
plt.show()
```
