#### 워드 클라우드

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
# %matplotlib inline
plt.rcParams['font.family'] = 'Malgun Gothic'
```

```python
tmp='강아지 산책 강아지 목욕 강아지 미용 강아지 쇼핑 친구와 저녁 먹음 가족과 점심 먹음 혼자 저녁 먹음 친구와 쇼핑'
```

```python
wc=WordCloud(font_path="C:/windows/fonts/malgun.ttf",
             background_color='black',
             width=500,
             height=500,
             max_words=10,
             max_font_size=200)
wc=wc.generate(tmp)
wc.to_image()
```

```python
print(wc.words_)
```

```python
plt.figure(figsize=(5,5))
plt.imshow(wc)
plt.axis('off')
plt.show()
```

```python
f=open('빅카인즈_코로나.txt', 'r', encoding='utf-8')
txt=f.read()
f.close()

wc=WordCloud(font_path="C:/windows/fonts/malgun.ttf",
             background_color='black',
             width=500,height=500,
             max_words=100,
             max_font_size=100,
            min_font_size=10)
wordFre=wc.generate(txt)

wordFre.to_image()
```
